```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
Bank_df = pd.read_csv('DS1_C8_V3_ND_Sprint3_BankChurn_Data.csv')
Bank_df

# Reading the csv file in to jupiter note book using pd.read_csv
```

```python
Bank_df.describe()

# by using the describe () we get the statistics of the data
```

```python
plt.figure(figsize=(30,10))
sb.boxplot(data=Bank_df)
plt.yscale('log')
plt.show()
```

```python
mean = round(np.mean(Bank_df),2)
mean
```

```python
std_dev = np.std(Bank_df)
std_dev
# Finding the standard deviation
```

```python
k = 3
lower_bound = mean - k * std_dev
upper_bound = mean + k * std_dev

outliers = (Bank_df < lower_bound) | (Bank_df > upper_bound)
# usinh k-sigma methof finfing the outliers
```

```python
Bank_df.columns[outliers.any()]
```

```python
rows_with_outliers = outliers[['Education_Level', 'Income_Category', 'Card_Category',
       'Contacts_Count_12_mon', 'Total_Trans_Ct', 'Avg_Utilization_Ratio']].any(axis=1)

# .any(axis=1) is checking if any of the values in each row (axis=1) of the selected columns are considered True 
# The result of this operation will be a boolean Series with True or False values for each row
# rows_with_outliers will shou us that whether each row in the selected columns contains at least one outlier.
```

```python
Bank_clean_data = Bank_df[~rows_with_outliers]
# Storing the data with no outliers in a new data frame
```

```python
Bank_clean_data
```

```python
Bank_clean_data.info()
# checking the data type of columns
```

```python
Bank_clean_data.isnull().sum()
# checking if there are any null values in the dataframe
```

```python
mode_card_category = Bank_clean_data['Card_Category'].mode()[0]
Bank_clean_data['Card_Category'].fillna(mode_card_category,inplace=True)

mean_credit_limit = Bank_clean_data['Credit_Limit'].mean()
Bank_clean_data['Credit_Limit'].fillna(mean_credit_limit,inplace=True)
# using .fillna() filling the missing values based on their data types
# For categorical data 'Card_Category' filling the null values with mode
# For numerical data 'Credit_Limit' filling the null values with mean
```

```python
Bank_clean_data.to_csv('Bank_clean_data.csv',index=False)
```
